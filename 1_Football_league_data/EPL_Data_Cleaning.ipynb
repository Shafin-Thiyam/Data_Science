```python
def Goals_scored(Teams,Dataframe):
    HG=[]
    AG=[]
    for team in Teams:        
        HG.append(np.sum(Dataframe[Dataframe.HomeTeam==team].FTHG))
        AG.append(np.sum(Dataframe[Dataframe.AwayTeam==team].FTAG))
    Home_Goals=pd.Series(HG)
    Away_Scored=pd.Series(AG)
    Goal_From=pd.Series(Home_Goals+Away_Scored)
    return Goal_From

def Goals_Incurred(Teams,Dataframe):
    HG=[]
    AG=[]
    for team in Teams:
        HG.append(np.sum(Dataframe[Dataframe.HomeTeam==team].FTAG))
        AG.append(np.sum(Dataframe[Dataframe.AwayTeam==team].FTHG))
    Home_Goals=pd.Series(HG)
    Away_Goals=pd.Series(AG)
    Goal_Against=pd.Series(Home_Goals+Away_Goals)
    return Goal_Against
```

```python
def Home_result_count(Dataset, Teams,Result):
    Home_win=[]
    Home_Lose=[]
    Home_Draw=[]
    for team in Teams:
        for R in Result:
            if R=='H':
                Home_win.append(len(Dataset[(Dataset.HomeTeam==team) & (Dataset.FTR==R)].FTR))
            elif R=='A':
                Home_Lose.append(len(Dataset[(Dataset.HomeTeam==team) & (Dataset.FTR==R)].FTR))
            else:
                Home_Draw.append(len(Dataset[(Dataset.HomeTeam==team) & (Dataset.FTR==R)].FTR))
    return [Home_win,Home_Lose,Home_Draw]

def Away_result_count(Dataset, Teams,Result):
    Away_win=[]
    Away_Lose=[]
    Away_Draw=[]
    for team in Teams:
        for R in Result:
            if R=='A':
                Away_win.append(len(Dataset[(Dataset.AwayTeam==team) & (Dataset.FTR==R)].FTR))
            elif R=='H':
                Away_Lose.append(len(Dataset[(Dataset.AwayTeam==team) & (Dataset.FTR==R)].FTR))
            else:
                Away_Draw.append(len(Dataset[(Dataset.AwayTeam==team) & (Dataset.FTR==R)].FTR))
    return [Away_win,Away_Lose,Away_Draw]
```

```python
def Shots_Corner(Dataset, Teams):
    S=[]
    ST=[]
    C=[]
    for team in Teams:
        S.append(np.sum(Dataset[(Dataset.HomeTeam==team)].HS) + np.sum(Dataset[(Dataset.AwayTeam==team)].AS))
        ST.append(np.sum(Dataset[(Dataset.HomeTeam==team)].HST) + np.sum(Dataset[(Dataset.AwayTeam==team)].AST))
        C.append(np.sum(Dataset[(Dataset.HomeTeam==team)].HC) + np.sum(Dataset[(Dataset.AwayTeam==team)].AC))
    return [pd.Series(S), pd.Series(ST) ,pd.Series(C)]
```

```python

def Foul(Dataset,Teams):
    F=[]
    Y=[]
    R=[]
    for team in Teams:
        F.append(np.sum(Dataset[(Dataset.HomeTeam==team)].HF) + np.sum(Dataset[(Dataset.AwayTeam==team)].AF))
        Y.append(np.sum(Dataset[(Dataset.HomeTeam==team)].HY) + np.sum(Dataset[(Dataset.AwayTeam==team)].AY))
        R.append(np.sum(Dataset[(Dataset.HomeTeam==team)].HR) + np.sum(Dataset[(Dataset.AwayTeam==team)].AR))
    return [pd.Series(F), pd.Series(Y), pd.Series(R)]
    
```

```python
def point_calculator(Dataset):
    points =pd.Series(Dataset.Total_Draw+(Dataset.Total_Win*3))
    return points
```
