```python
#Prs_AgeNSex_CAW_2012=pd.read_csv(women_Dts_dir+'Police_Arrest_of_Age_sex__CrimeAgainsWomen_2012.csv')
Rankwise_Police_strength_2001_2012=pd.read_csv(women_Dts_dir+"Rankwise_Police_strength_2001-2012.csv")
Rankwise_Police_strength_2001_2012.columns=map(str.lower,Rankwise_Police_strength_2001_2012.columns)

Rankwise_Police_strength_2013=pd.read_csv(women_Dts_dir+"Rankwise_Police_strength_2013.csv")
Rankwise_Police_strength_2013.columns=map(str.lower,Rankwise_Police_strength_2013.columns)
```

```python
Rankwise_Police_strength_2001_2012['state/ut']=Rankwise_Police_strength_2001_2012['state/ut'].str.upper()
Rankwise_Police_strength_2013['state/ut']=Rankwise_Police_strength_2013['state/ut'].str.upper()
```

```python
Rankwise_Police_strength_2001_2012
```

```python
Rankwise_Police_strength=pd.merge_ordered(Rankwise_Police_strength_2001_2012,Rankwise_Police_strength_2013)
Rankwise_Police_strength.columns=['state/ut','year','dg/adg','ig','dig','ssp/sp/ad.sp','asp/dysp','insepctor','sub_inspector','asst.sub_inspector','head constable','below_hc_above_constable','constables','total']
```
